```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel("/workspaces/codespaces-jupyter/Practice-Machine-Learning/data1.xlsx")
```

```python
#formula price = m * area + b
```

```python
plt.xlabel("area",fontsize = 22)
plt.ylabel("price",fontsize = 22)
plt.scatter(df.area,df.price,color = "red",marker = "+")
plt.plot(df[['area']],df[['price']],color = 'blue')
```

```python
regresion = linear_model.LinearRegression().fit(df[['area']],df.price)
```

```python
regresion.predict([[3300]])
```

```python
df2 = pd.read_excel("/workspaces/codespaces-jupyter/Practice-Machine-Learning/data4.xlsx")
regresion.fit(df[['area']],df[['price']])
df2['price'] = regresion.predict(df2)
```

```python
%matplotlib inline
plt.xlabel('area',fontsize = 22)
plt.ylabel('ylabel',fontsize = 22)
plt.scatter(df2[['area']],df2[['price']],color = 'red', marker = "+")
plt.plot(df2[['area']],df2[['price']], color = "blue")
```
